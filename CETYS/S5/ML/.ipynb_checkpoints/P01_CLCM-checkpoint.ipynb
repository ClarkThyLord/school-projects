1.5.Utilizar el vector ð‘¥ creado en la actividad 1.4 y evaluar las siguientes funciones:1)ð‘¦=sinð‘¥,2) ð‘¦=log10ð‘¥,3) ð‘¦=ð‘’ð‘¥.Despliegue en pantalla el resultado numÃ©rico y grÃ¡fico.

```python
x = [x * 0.1 for x in range(-20, 20)]

func_1 = [math.sin(n) for n in x]
func_2 = [math.log(abs(n)) if n > 0 else None for n in x]
func_3 = [math.e * n for n in x]

plt.title("1.5 func_1(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, func_1)
plt.show()
print("func_1 = ", func_1)

plt.title("1.5 func_2(x) = log(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, func_2)
plt.show()
print("func_2 = ", func_2)

plt.title("1.5 func_3(x) = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, func_3)
plt.show()
print("func_3 = ", func_3)
```

1.9. Realizar un programa que despliegue los siguientes patrones utilizando asteriscos.

```python
size = 0
while True:
    try:
        size = int(input("TamaÃ±o de Patrones: "))
        if size > 3 and size < 11: break
    except ValueError: pass

result = []
for l in range(size):
    result.append("")
    for c in range(size):
        result[l] += "*" if c - 1 < l else " "

p = ""
for l in result:
    p += l + "\n"
print(p)
p = ""
for l in result:
    p += l[::-1] + "\n"
print(p)

p = ""
result = result[::-1]
for l in result:
    p += l + "\n"
print(p)
p = ""
for l in result:
    p += l[::-1] + "\n"
print(p)
```
